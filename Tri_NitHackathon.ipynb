```python
import sklearn
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
# Create the dataframe
df_boston = data
df_boston.columns = df_boston.columns
df_boston.head()
 
''' Detection '''
# IQR
Q1 = np.percentile(df_boston['rainfall'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df_boston['rainfall'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", df_boston.shape)
 
# Upper bound
upper = np.where(df_boston['rainfall'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df_boston['rainfall'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
df_boston.drop(upper[0], inplace = True)
df_boston.drop(lower[0], inplace = True)
 
print("New Shape: ", df_boston.shape)
```

```python
data = df_boston
```

```python
X = data.drop('label',axis=1)
y = data['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.30,
                                                    shuffle=True,random_state=0)
```

```python
import lightgbm as lgb
model = lgb.LGBMClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_pred,y_test)
print('LightGBM Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
def create_2d_array(N, P, K, temperature, humidity, pH, rainfall):
    data = []
    for i in range(N):
        sub_list = [P, K, temperature, humidity, pH, rainfall]
        data.append(sub_list)
    return data

N = int(input("Enter number of rows: "))
P = int(input("Enter value of P: "))
K = int(input("Enter value of K: "))
temperature = int(input("Enter value of temperature: "))
humidity = int(input("Enter value of humidity: "))
pH = float(input("Enter value of pH: "))
rainfall = int(input("Enter value of rainfall: "))

result = create_2d_array(N, P, K, temperature, humidity, pH, rainfall)
print(result)
```
